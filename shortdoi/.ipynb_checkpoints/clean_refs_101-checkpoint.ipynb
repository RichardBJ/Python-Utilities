```python
import re
```

```python
def do_one_line(line):
        '''m = re.search("\,.*?\(",line)'''
        m = re.search("\,.*? \((?=\d\d\d\d)",line)    
        substring=m.group(0)
        newline= line.replace(substring," (")
        newline= newline.replace("). h",") h")
        return newline
```

```python
'''Dont let it end with a blank line obs'''
with open('references.txt', 'rt') as filein:
    for line in filein:
        print(line)
        newline = do_one_line(line)
        print(newline)
        with open('minirefs.txt','a') as fileout:
            fileout.write(newline)
            
    
```
