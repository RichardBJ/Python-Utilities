```python
def do_one_doi(url):
    print("\n"+url)
    requestioobject=requests.get(url=url)
    data=requestioobject.json()
    shortDOI="https://doi.org/"+data["ShortDOI"]
    shortDOI=shortDOI.replace(r"/10/",r"/")
    return shortDOI
    
```

```python
def del_doi(line):
    x = line.rpartition("http")
    return x[0]
```

```python
def do_one_line(line,DELETE_ORIGINAL_DOI=False):
        with open('shortDOI_2.txt', 'a') as file:
            if "doi.org" in line:
                result = re.search('https://doi.org:(.*)', line)
                doi = result.group(1)
                print(doi)
                file.write("\n\n")
                if DELETE_ORIGINAL_DOI:
                    line=del_doi(line)
                file.write(line)
                url = "http://shortdoi.org/"+doi+"?format=json"
                if not DELETE_ORIGINAL_DOI:
                    file.write(url)
                shortDOI=do_one_doi(url)
                print(shortDOI)
                file.write("\n")
                file.write(shortDOI)
            else:
                file.write("\n\n")
                file.write(line)
                



    
```

```python
DELETE_ORIGINAL_DOI=True

file = open('minirefs.txt', 'rt')
for line in file:
    print(line)
    do_one_line(line, DELETE_ORIGINAL_DOI)
```
